```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams

def plot_excel_data(excel_file, my_title, x_label, y_label):
    # 读取Excel文件
    df = pd.read_excel(excel_file, index_col=0)

    

    # 设置字体
    rcParams['font.family'] = 'sans-serif'
    rcParams['font.sans-serif'] = ['Arial']
    # 设置seaborn主题
    sns.set_theme(style="white", palette=None)  # 选择其他主题
    # 创建绘图
    plt.figure(figsize=(8, 6))
    sns.lineplot(x=df.iloc[:, 3], y=df.iloc[:, 0], marker='o', linestyle='dashed', color='b')

    # 设置标题和标签
    plt.title(my_title)
    plt.xlabel(x_label, fontsize=20)
    plt.ylabel(y_label, fontsize=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.savefig('tda_data_model_train_acc.pdf')

    # 显示图例
    # plt.legend()
    

    # 显示图表
    plt.show()

# 示例用法
plot_excel_data('Acc_TDA_Gap.xlsx', None,'Gaps in TDA (total survival time)', 'Training accuracy of models ')
```
