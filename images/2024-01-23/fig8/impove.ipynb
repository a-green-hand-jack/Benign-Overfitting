```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_excel_data_as_bar(file_path, sheet_name):
    """
    从Excel文件中读取数据，按照行分组，绘制柱状图。

    参数:
    - file_path: Excel文件路径。
    - sheet_name: 指定的Sheet名称。
    """
    # 读取Excel文件
    df = pd.read_excel(file_path, sheet_name=sheet_name, index_col=0)

    # 获取行名和列名
    row_names = df.index
    column_names = df.columns

    # 设置颜色和图例
    colors = ['r', 'orange', 'green']
    legends = ['original', 'rk', 'S']
    legends2 = ['original', r'$r_{k^*}$', '$S$']

    # 绘制柱状图
    width = 0.25  # 每个柱的宽度
    x = range(len(row_names))
    for i, legend in enumerate(legends):
        plt.bar(
            [pos + width * i for pos in x],  # x轴位置，分组展示
            df[legend],  # 柱的高度
            width=width,  # 柱的宽度
            label=legend,  # 柱的图例
            color=colors[i]  # 柱的颜色
        )

    # 添加标题和轴标签
    plt.title(f'Improvements by {sheet_name}', fontsize=30)  # 调整标题字体大小
    plt.ylabel('mIoU', fontsize=25)  # 调整y轴标签字体大小

    # 设置图例
    plt.legend(legends2, fontsize=18)

    # 设置x轴标签
    plt.xticks([pos + width for pos in x], row_names, fontsize=25)  # 调整x轴标签字体大小

    # 调整图表尺寸
    plt.gcf().set_size_inches(8, 6)  # 调整图表尺寸为8x6英寸

    # 调整布局，防止x轴标签被截断
    plt.tight_layout()
    plt.savefig('unext_medical.pdf')
    # 显示图表
    plt.show()

# 示例用法，将'./your_file.xlsx'替换为实际的文件路径
plot_excel_data_as_bar('./impove.xlsx', 'crop')
```
