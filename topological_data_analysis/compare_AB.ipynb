# 比较不同的after_betti_bars

具体的讲，就是要比较原始的`data`,`LaNet`,`MLP`的输出的after betti 的四个维度
1. bars的总数目【所有的ε对应的bars的数目的总和】
2. bars的总存活时间【所有的bar对应的ε的长度（生死长度）的总和】
3. Max(bar(ε))【考察不同ε对应的bars的数目，选择bars数目最多的εs中的最大的ε】
4. bars的边缘长度【固定两个bar之间的距离是0.1，在这个基础上计算每一个图像下的bar的生线和死线的长度】

但是，现在这些数据都是孤立的保存在对应的文件夹下面，还没有手段在它们之间进行操作。

也就是：
* 还没办法得到原始数据和不同网络之间的某一个增强下的after betti 的差距；
* 也没有办法得到同一个网络或者data下的同一种增强的不同强度之间的规律。

# 比较after betti number 数据

也就是读取每一个**同一中增强类型下的不同的网络和数据**产生一的`compare_different_bitte_norm_in_same_augmentation.pkl`文件

```python
import os
from dataset.after_betti import custom_sort
from pprint import pprint
import pickle

def get_all_cabs(base_path):
    parent_files = os.listdir(base_path)
    parent_files.sort(key=custom_sort)
    result_3d_dfs = {}
    for child in parent_files:
        child_path = os.path.join(base_path, child)
        # 判断路径是否是一个文件夹
        if os.path.isdir(child_path):
            # print(child_path)
            child_files = os.listdir(child_path)
            child_files.sort(key=custom_sort)
            for grandchild in child_files:
                grandchild_path = os.path.join(child_path, grandchild)
                if os.path.isdir(grandchild_path):
                #    print(grandchild_path)
                    # compare_after_betti_in_same_augmentation(grandchild_path)
                    after_betti_data_path = os.path.join(grandchild_path, "compare_different_bitte_norm_in_same_augmentation.pkl")
                    print(after_betti_data_path)
                    with open(after_betti_data_path, 'rb') as file:
                        after_betti_data = pickle.load(file)
                        pprint(after_betti_data)

k = get_all_cabs(".\\distance")
```
